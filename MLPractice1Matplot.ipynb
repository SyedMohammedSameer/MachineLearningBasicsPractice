**A Quick Tutorial on Matplotlib**

**Why Data Visualisation?**

Human Brain processes data easily when its pictorial or graphical form
It allows quickly to interpret the data 

**What is Matplotlib?**

A python package used for 2D graphics

**Types of plot**

1. Bar Graph
2. Histogram
3. Scatter plot
4. Pie plot
5. Hexagonal Bin Plot
6. Area Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#plotting values 1D
a=[1,3,4,5]
plt.plot(a)
plt.show
```

```python
# Plotting 2D values
x=[3,4,5,7]
y=[1,2,6,8]
plt.plot(x,y)
plt.show()
```

```python
#Scatter plot of 2D values
x=[3,4,5,7]
y=[1,2,6,8]
plt.scatter(x,y)
plt.show()
```

```python
#Scatter Plot with random values
N=43
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=np.pi*(15*np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors,alpha=0.7)
plt.show
```

```python
m=20
b=2
x=np.array([1,8,23,24])
y=m*x+b
plt.scatter(x,y)
plt.show()
```

```python
N=43
x=np.array([4,2,8,6])
y=np.array([1,3,7,8])
plt.plot(x,y,color='Cyan',marker='o')
plt.show()
```

```python
N=43
x=np.array([4,2,8,6])
y=np.array([1,3,7,8])
plt.plot(x,y,'r--',linewidth=3)
plt.show()
```

```python
N=43
x=np.array([4,2,8,6])
y=np.array([1,3,7,8])
plt.plot(x,y)
plt.ylabel("Random Values")
plt.xlabel("Pointers")
plt.title("Plotting a random graph")
plt.show()
```

```python
N=43
x=np.array([4,2,8,6])
y=np.array([1,3,7,8])
plt.plot(x,y,'r--',linewidth=3)
plt.axis([0,6,0,20])
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,70,30,20])
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,70,30,20])
plt.xscale('log')
plt.yscale('linear')
plt.plot(x,y)
plt.show()
```

```python
#Pie Plot
labels='India','Ireland','UK','Germany'
sizes=[38,32,20,10]
explode=(0.2,0.1,0,0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
#Histogram
mu=100
sigma=15
x=mu+sigma*np.random.randn(10000)
n,bins,patches=plt.hist(x,50,normed=1,facecolor='blue',alpha=0.80)
plt.xlabel('Smart Graphs')
plt.ylabel('Probability')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
days=[1,2,3,4,5]
sleeping=[8,6,7,5,8]
eating=[2,3,2,4,3]
working=[9,11,9,10,12]
playing=[5,4,6,5,1]
```

```python
#Stack Plot
plt.plot([],[],color='m',label='Sleeping')
plt.plot([],[],color='c',label='Eating')
plt.plot([],[],color='r',label='Working')
plt.plot([],[],color='k',label='Playing')
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as pit

slices = [7,3,7,7]

activities = ['sleeping', 'eating', 'working','playing']

cols = ['c','m', 'r','b']

plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow= True,explode=(0,0.1,0,0), autopct='%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
